## Line Plots

```julia
using Plots
x = 1:10; y = rand(10); # These are the plotting data
plot(x, y)
```

Every column is a *series*, i.e. a set of related points which form lines, surfaces etc. <br>
Thus we can plot multiple lines by plotting a matrix of values and each column is interpreted as a separate line:

```julia
using Plots
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y)
```

```julia
## Mutation
z = rand(10)
plot!(x, z)
```

```julia
## Explicit
x = 1:10; y = rand(10, 2) # 2 columns means two lines
p = plot(x, y)
z = rand(10)
plot!(p, x, z)
```

## Attributes: The plot styler

Note:  positional arguments are input data, and keyword arguments are attributes. Thus something like `plot(x,y,z)` is 3-dimensional data for 3D plots, while `plot(x,y,attribute=value)` is 2-dimensional with an attribute.

Attrubutes in the following cell:
1. title: title of the plot 
2. label: for legend's label 
3. line width (lw)
4. xlabel
5. ylabel 

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y, title = "Two Lines", label = ["Bluee" "Orangee"], lw = 3)
```

 Attribute can also be applied by mutating the plot with a modifier function <br>
 Note: this implicitly uses the global `Plots.CURRENT_PLOT` and we can apply it to other plot objects via `attribute!(p,value)`. 

```julia
xlabel!("My x label")
```

## Plotting Functions

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```
